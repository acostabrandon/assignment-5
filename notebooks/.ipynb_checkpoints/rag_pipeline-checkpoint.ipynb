## 2. Load Job Descriptions and Resume

**Required:**
- 10+ JD files in `data/job_descriptions/` (each as a separate .txt or .pdf)
- Your resume in `data/resume/`
- A metadata file `data/jd_metadata.csv` with columns: filename, company, title, source_url, date_collected

Print: number of JDs loaded, number of resume docs, and preview content from each.

```python
# ── Load JD metadata ──

from pathlib import Path
import pandas as pd

# Define project paths
DATA_DIR = Path("../data")
JD_DIR = DATA_DIR / "job_descriptions"
RESUME_DIR = DATA_DIR / "resume"
METADATA_PATH = DATA_DIR / "jd_metadata.csv"

# Load metadata CSV
jd_metadata = pd.read_csv(METADATA_PATH)

# Preview metadata
print("JD metadata loaded successfully.")
print("Number of rows in metadata:", len(jd_metadata))

display(jd_metadata.head(10))
```

```python
# ── Load JD documents and resume ──

jd_documents = []
resume_documents = []

# Load job description text files
for _, row in jd_metadata.iterrows():
    filename = row["filename"]
    file_path = JD_DIR / filename
    
    if file_path.exists():
        text = file_path.read_text(encoding="utf-8")
        
        doc = Document(
            page_content=text,
            metadata={
                "doc_type": "job_description",
                "filename": filename,
                "company": row["company"],
                "title": row["title"],
                "source_url": row["source_url"],
                "date_collected": row["date_collected"]
            }
        )
        
        jd_documents.append(doc)
    
    else:
        print(f"Missing file: {file_path}")

# Load resume text file
resume_files = list(RESUME_DIR.glob("*.txt"))

if len(resume_files) == 0:
    print("No resume .txt file found in data/resume/")
else:
    resume_path = resume_files[0]
    resume_text = resume_path.read_text(encoding="utf-8")
    
    resume_doc = Document(
        page_content=resume_text,
        metadata={
            "doc_type": "resume",
            "filename": resume_path.name
        }
    )
    
    resume_documents.append(resume_doc)

print("Job descriptions loaded:", len(jd_documents))
print("Resume documents loaded:", len(resume_documents))
```

```python
# ── Preview sample content ──

print("=== Sample Job Description ===")
print("Company:", jd_documents[0].metadata["company"])
print("Title:", jd_documents[0].metadata["title"])
print("Filename:", jd_documents[0].metadata["filename"])
print()
print(jd_documents[0].page_content[:1000])

print("\n" + "=" * 80 + "\n")

print("=== Resume Preview ===")
print("Filename:", resume_documents[0].metadata["filename"])
print()
print(resume_documents[0].page_content[:1000])
```

## 3. Text Chunking

Split your documents into chunks.  
**Required:** Try at least 2 chunking strategies, compare them quantitatively, and justify your final choice.

**Hint:** JDs often have natural sections (Requirements, Responsibilities, Qualifications). Consider whether your splitter respects these boundaries.

```python
# ── Strategy 1 ──

# Strategy 1: Larger recursive chunks
# This keeps more surrounding context together, but may mix multiple JD sections in one chunk.

strategy_1_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    separators=["\n\n", "\n", ". ", " ", ""]
)

all_documents = jd_documents + resume_documents

strategy_1_chunks = strategy_1_splitter.split_documents(all_documents)

print("Strategy 1 complete.")
print("Total documents before chunking:", len(all_documents))
print("Total chunks created:", len(strategy_1_chunks))

# Count chunks by document type
strategy_1_counts = {}

for chunk in strategy_1_chunks:
    doc_type = chunk.metadata.get("doc_type", "unknown")
    
    if doc_type not in strategy_1_counts:
        strategy_1_counts[doc_type] = 0
    
    strategy_1_counts[doc_type] += 1

print("Chunks by document type:")
print(strategy_1_counts)

# Preview one chunk
print("\nSample chunk preview:")
print(strategy_1_chunks[0].page_content[:1000])
```

```python
# ── Strategy 2 ──

# Strategy 2: Smaller section-aware recursive chunks
# This is designed for job descriptions because JDs often have sections like:
# Responsibilities, Qualifications, Requirements, Skills, and Preferred Qualifications.

strategy_2_splitter = RecursiveCharacterTextSplitter(
    chunk_size=600,
    chunk_overlap=100,
    separators=[
        "\n\nResponsibilities",
        "\n\nQualifications",
        "\n\nRequirements",
        "\n\nPreferred Qualifications",
        "\n\nSkills",
        "\n\nAbout",
        "\n\n",
        "\n",
        ". ",
        " ",
        ""
    ]
)

strategy_2_chunks = strategy_2_splitter.split_documents(all_documents)

print("Strategy 2 complete.")
print("Total documents before chunking:", len(all_documents))
print("Total chunks created:", len(strategy_2_chunks))

# Count chunks by document type
strategy_2_counts = {}

for chunk in strategy_2_chunks:
    doc_type = chunk.metadata.get("doc_type", "unknown")
    
    if doc_type not in strategy_2_counts:
        strategy_2_counts[doc_type] = 0
    
    strategy_2_counts[doc_type] += 1

print("Chunks by document type:")
print(strategy_2_counts)

# Preview one chunk
print("\nSample chunk preview:")
print(strategy_2_chunks[0].page_content[:1000])
```

```python
# ── Compare strategies ──

def summarize_chunks(chunks, strategy_name):
    chunk_lengths = [len(chunk.page_content) for chunk in chunks]
    
    summary = {
        "strategy": strategy_name,
        "total_chunks": len(chunks),
        "avg_chunk_length": round(sum(chunk_lengths) / len(chunk_lengths), 2),
        "min_chunk_length": min(chunk_lengths),
        "max_chunk_length": max(chunk_lengths),
        "jd_chunks": sum(1 for chunk in chunks if chunk.metadata.get("doc_type") == "job_description"),
        "resume_chunks": sum(1 for chunk in chunks if chunk.metadata.get("doc_type") == "resume")
    }
    
    return summary

chunking_comparison = pd.DataFrame([
    summarize_chunks(strategy_1_chunks, "Strategy 1: Larger recursive chunks"),
    summarize_chunks(strategy_2_chunks, "Strategy 2: Smaller section-aware chunks")
])

display(chunking_comparison)
```

### Chunking Decision

**Which strategy did you choose?**  I selected Strategy 2 as the final chunking approach. 

**Why?**  Streagy 2 used smaller, section-aware recursive chunks with a chunk size of 600 and an overlap of 100. This strategy is better suited for job descriptions because many postings are organized into sections such as responsibilities, qualifications, requirements, preferred qualifications, and skills. Smaller chunks make retrieval more focused and easier to compare with evidence from the resume. Precise skill gaps, keyword alignment, and fit summary analysis are all stronger with improved and more specific chunking. 

**Final settings (chunk_size, overlap):** Smaller section-aware chunks, max_chunk_length = 600, overlap = 100

## 5. Analysis Prompts and Chain

Build 3 analysis types, each with its own prompt:

1. **Skill Gap Report:** Compare resume skills vs. JD requirements. Output matching skills, missing skills, and recommended actions.
2. **Keyword Alignment:** Extract key terms from a JD, check which appear in the resume, report a match rate.
3. **Fit Summary:** 3-4 sentence narrative assessment citing evidence from both documents.

You also need to wire up the LLM and a way to pass a specific JD + resume into each prompt.

**Required:** Document at least 3 prompt iterations total (across any analysis type) with rationale.

**Reminder:** Prompt design must be your own work (Tier 2 — AI prohibited for this step).

```python
# ── Initialize LLM ──

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
    api_key=openai_api_key
)

print("LLM initialized successfully.")
print("Model: gpt-4o-mini")
print("Temperature: 0")
```

```python
# ── Analysis 1: Skill Gap Report ──
# — Analysis 1: Skill Gap Report —

# Choose one JD to test
selected_jd_filename = "jd_01_sales_operations_analyst.txt"

def get_jd_context(filename, query, k=5):
    """
    Retrieve relevant chunks from one selected job description.
    """
    results = vector_store.similarity_search(
        query=query,
        k=k,
        filter={
            "filename": filename
        }
    )
    
    return "\n\n".join([doc.page_content for doc in results])


def get_resume_context(query, k=5):
    """
    Retrieve relevant chunks from the resume.
    """
    results = vector_store.similarity_search(
        query=query,
        k=k,
        filter={
            "doc_type": "resume"
        }
    )
    
    return "\n\n".join([doc.page_content for doc in results])


skill_gap_jd_context = get_jd_context(
    filename=selected_jd_filename,
    query="required skills qualifications responsibilities experience tools",
    k=5
)

skill_gap_resume_context = get_resume_context(
    query="skills experience projects tools education responsibilities",
    k=5
)

skill_gap_prompt_text = """
Pretend you are a recruiter reviewing job applications using a resume and job description. You are being provided the job description text file and resume text file. Please compare the job description's requirements with the resume's skill section. 

Your output must include (1) a list of matching skills, (2) a list of missing skills, and (3) recommended actions to close the skill gap. 

You must only use the provided inputs. Do not hallucinate experience, make assumptions, or make changes to the input files.

Job Description: {job_description}
Resume: {resume}
"""

skill_gap_prompt = PromptTemplate(
    input_variables=["job_description", "resume"],
    template=skill_gap_prompt_text
)

skill_gap_chain = skill_gap_prompt | llm | StrOutputParser()

skill_gap_result = skill_gap_chain.invoke({
    "job_description": skill_gap_jd_context,
    "resume": skill_gap_resume_context
})

print(skill_gap_result)
```

```python
# ── Analysis 2: Keyword Alignment ──

keyword_jd_context = get_jd_context(
    filename=selected_jd_filename,
    query="keywords tools technologies qualifications required skills preferred skills",
    k=5
)

keyword_resume_context = get_resume_context(
    query="skills tools technologies software platforms certifications projects experience",
    k=5
)

keyword_alignment_prompt_text = """
Pretend you are a recruiter reviewing resumes to match them to a job description. You are being provided with the job description and resume contexts for review. 

Job Description: {job_description}
Resume: {resume}

Your output will be a table. Column1 of the table will be a list of keywords or phrases directly from the job description. These should be justified, specific, and not repeated. Keep them brief, big-picture keywords from the job description. 

Column 2 will indicate whether each keyword matches the provided resume. If each keyword directly matches the resume, it is a match. If not, it is missing, or requires expansion. This is a Match/No Match result.

Column 3 should serve as your justification from the resume for the match, to ensure that it is proper. For any slight matches, you can include your justification for decision in this column as well. Keep your reasoning concise. 
Below the table, provide a final match rate for the keywords between the job description and resume. 

Do not hallucinate any keywords or skills from either the job description or resume; you are only limited to those. Do not make assumptions. 
"""

keyword_alignment_prompt = PromptTemplate(
    input_variables=["job_description", "resume"],
    template=keyword_alignment_prompt_text
)

keyword_alignment_chain = keyword_alignment_prompt | llm | StrOutputParser()

keyword_alignment_result = keyword_alignment_chain.invoke({
    "job_description": keyword_jd_context,
    "resume": keyword_resume_context
})

print(keyword_alignment_result)
```

```python
# ── Analysis 3: Fit Summary ──

fit_jd_context = get_jd_context(
    filename=selected_jd_filename,
    query="role responsibilities qualifications required experience preferred qualifications",
    k=5
)

fit_resume_context = get_resume_context(
    query="professional experience education skills projects achievements tools",
    k=5
)

fit_summary_prompt_text = """
Pretend you are a recruiter reviewing resumes and matching them against the job description for a job posting. You are being provided the resume context and job description context. 

Job Description: {job_description} 
Resume: {resume} 

Do a comprehensive analysis of the keywords, experience requirements, and desired skill sets in the job description. Match this against the provided resume. 

Write a short 3-4 sentence narrative assessment of the job fit, make sure it is no longer than 4 sentences. 

Make sure to cite evidence from both the job description and resume. Do not make hallucinations, assumptions from the resume, or exaggerate the fit. Make sure there is no bias. 
"""

fit_summary_prompt = PromptTemplate(
    input_variables=["job_description", "resume"],
    template=fit_summary_prompt_text
)

fit_summary_chain = fit_summary_prompt | llm | StrOutputParser()

fit_summary_result = fit_summary_chain.invoke({
    "job_description": fit_jd_context,
    "resume": fit_resume_context
})

print(fit_summary_result)
```

Document at least 3 total iterations across any of the analysis types.

**Iteration 1:** For the second analysis, the first prompt worked very strongly:

Resume: {resume}_

_Please start by creating a table of keywords from the job description for your output. Then, compare each keyword to the resume provided, and rank whether or not they are matched._

However, the output ranked each match as "Strong", "Moderate", or "Weak", which is objective but does not provide sufficient analysis. **Iteration 2:** For the second analysis, the second prompt made significant improvements from the previous iteration:

_Pretend you are a recruiter reviewing resumes to match them to a job description. You are being provided with the job description and resume contexts for review._

_Job Description: {job_description}
Resume: {resume}_

_Your output will be a table. Column1 of the table will be a list of keywords or phrases directly from the job description. These should be justified, specific, and not repeated. Keep them brief, big-picture keywords from the job description._ 

_Column 2 will indicate whether each keyword matches the provided resume. If each keyword directly matches the resume, it is a match. If not, it is missing, or requires expansion. This is a Match/No Match result._

This prompt helped simplify the output by specifically requesting concise reasoning and development of the keywords/phrases list. Furthermore, this prompt removes the potential for any hallucinations or assumptions. It also uses more concise language to justify the match and restricts the model to a binary selection between Match and No Match. Furthermore, it removes the complicated calculation of the match score, and instead, just totals up the number of matches and no matches. 

**Iteration 3:** For the third analysis, the second iteration of the prompt was significantly more optimized:

The original prompt lacked a specific request for evidence to justify the job fit. Asking it to cite specific evidence helps ensure that no misrepresentations, assumptions, or exaggerations are made. The model is also restricted by a specific request to prevent hallucinations. Because this is a broader prompt, asking it to be concise and evidence-based helps prevent mistakes.

## 6. Zero-shot vs. Few-shot Comparison

Pick one of your 3 analysis types. Create a few-shot version by adding 1-2 example input/output pairs to the prompt. Run both versions on the same JD and compare outputs.

**Reminder:** You must write the few-shot examples yourself (Tier 2).

```python
# ── Few-shot version of your chosen analysis ──

# We are comparing the Skill Gap Report prompt from Step 5
# using the same selected JD and resume contexts.

few_shot_skill_gap_prompt_text = """
You are comparing a candidate's resume against a job description.

Use only the provided job description and resume. Do not make assumptions, exaggerations, or hallucinations.

Example 1:

Job Description:
The role requires SQL, dashboard development, stakeholder reporting, and experience translating business needs into analytics deliverables. This is for a Business Analyst role. 

Resume:
The candidate has built Tableau dashboards, used SQL for data extraction, and presented weekly KPI reporting to business stakeholders.

Your output must include (1) a list of matching skills, (2) a list of missing skills, and (3) recommended actions to close the skill gap. 

Example 2:

Job Description:
The role requires experience with Python, machine learning model evaluation, A/B testing, and cloud deployment. This is for a Marketing Analytics role.

Resume:
The candidate has Python coursework, built classification models in Jupyter Notebook, and evaluated models using accuracy and confusion matrices.

Your output must include (1) a list of matching skills, (2) a list of missing skills, and (3) recommended actions to close the skill gap. 

Job Description Context:
{job_description}

Resume Context:
{resume}
"""

few_shot_skill_gap_prompt = PromptTemplate(
    input_variables=["job_description", "resume"],
    template=few_shot_skill_gap_prompt_text
)

few_shot_skill_gap_chain = few_shot_skill_gap_prompt | llm | StrOutputParser()

few_shot_skill_gap_result = few_shot_skill_gap_chain.invoke({
    "job_description": skill_gap_jd_context,
    "resume": skill_gap_resume_context
})

print(few_shot_skill_gap_result)
```

```python
# — Run 9 analyses (3 JDs x 3 analysis types) —

top_3_rows = jd_metadata.head(3).copy()

analysis_outputs = []

for _, jd_row in top_3_rows.iterrows():
    
    jd_filename = jd_row["filename"]
    company = jd_row["company"]
    title = jd_row["title"]
    
    print(f"Running analyses for: {jd_filename}")
    
    # -----------------------------
    # 1. Skill Gap Report
    # -----------------------------
    skill_gap_jd_context = get_jd_context(
        filename=jd_filename,
        query="required skills qualifications responsibilities experience tools",
        k=5
    )
    
    skill_gap_resume_context = get_resume_context(
        query="skills experience projects tools education responsibilities",
        k=5
    )
    
    skill_gap_output = skill_gap_chain.invoke({
        "job_description": skill_gap_jd_context,
        "resume": skill_gap_resume_context
    })
    
    analysis_outputs.append({
        "jd_filename": jd_filename,
        "company": company,
        "title": title,
        "analysis_type": "Skill Gap Report",
        "jd_context_length": len(skill_gap_jd_context),
        "resume_context_length": len(skill_gap_resume_context),
        "output": skill_gap_output
    })
    
    # -----------------------------
    # 2. Keyword Alignment
    # -----------------------------
    keyword_jd_context = get_jd_context(
        filename=jd_filename,
        query="keywords tools technologies qualifications required skills preferred skills",
        k=5
    )
    
    keyword_resume_context = get_resume_context(
        query="skills tools technologies software platforms certifications projects experience",
        k=5
    )
    
    keyword_output = keyword_alignment_chain.invoke({
        "job_description": keyword_jd_context,
        "resume": keyword_resume_context
    })
    
    analysis_outputs.append({
        "jd_filename": jd_filename,
        "company": company,
        "title": title,
        "analysis_type": "Keyword Alignment",
        "jd_context_length": len(keyword_jd_context),
        "resume_context_length": len(keyword_resume_context),
        "output": keyword_output
    })
    
    # -----------------------------
    # 3. Fit Summary
    # -----------------------------
    fit_jd_context = get_jd_context(
        filename=jd_filename,
        query="role responsibilities qualifications required experience preferred qualifications",
        k=5
    )
    
    fit_resume_context = get_resume_context(
        query="professional experience education skills projects achievements tools",
        k=5
    )
    
    fit_output = fit_summary_chain.invoke({
        "job_description": fit_jd_context,
        "resume": fit_resume_context
    })
    
    analysis_outputs.append({
        "jd_filename": jd_filename,
        "company": company,
        "title": title,
        "analysis_type": "Fit Summary",
        "jd_context_length": len(fit_jd_context),
        "resume_context_length": len(fit_resume_context),
        "output": fit_output
    })

print("All analyses completed.")

analysis_outputs_df = pd.DataFrame(analysis_outputs)

display(analysis_outputs_df[[
    "jd_filename",
    "company",
    "title",
    "analysis_type",
    "jd_context_length",
    "resume_context_length"
]])
```

### Evaluation Analysis

**Which analysis type worked best?** 
The Fit Summary worked the best. Across all 3 job descriptions, the job fit summary was the only one that organized actionable insights by soft skills, technical skills, and industry experience. The fit summary had high accountability, was faithful throughout, and its list of skills was always correct, with no exaggerations or hallucinations. 

**Which JDs produced the best/worst results? Why?**
The third job description for the CRM Ops Associate yielded the best results across the 3 models. It was able to ignore the company description, job benefits, and recruitment timeline aspects of the job description, allowing for the list of skills and keyword to be most accurate. It also did an excellent job of segmenting education requirements, hard skills, and technical requirements, and the industry context and preferred professional experience in the JD. 

Contrastingly, the second job had the worst results for the opposite reasons. In the Keyword Alignment model, many of the skills were pulled from irrelevant sections of the job description. For instance, there was a retrieval of the recruitment notice section of the job description, legal notices, and very general soft skills like team player, hard-working mentality, etc. 

**Where did the system hallucinate or produce inaccurate results?**
Thankfully, the model did not provide any hallucinations thanks to the restrictions in the prompt engineering. That said, this served as a double edged sword, because a lot of the inaccurate results stemmed from the models inability to make logical reasonings from the resume to match the soft skills section of job descriptions. Nearly of the inaccurate results stemmed from very specific soft-skill listings like comments on employee "mindset", employee's "willingness to work long hours", or "maintain agility". Each job descriptions adjectives chosen for soft skills will vary significantly, thus, this caused the disconnect with the resume. Because these specific keywords were not in the resume, especially because thinks like problem-solving, collaboration, and work ethic can be inferred through professional experience, led to several no matches throughout the keyword alignment modeling. 

Furthermore, the innacurate results also seemed to stem from the education section. It seems that the resume only pulled from the professionla experience and skills section of the resume, not taking into account education. This led to a nearly every job description having the recommendation of getting a Bachelor's degree. Alternatively, the model may also lack the necessary udnerstanding of the B.A. abbreviation. 

**What would you improve?**
The first thing I would improve is a standardization of the job description input. This would require manual work, with the entry of the job description specifically removing legal hiring notices, mentions of job benefits or recruiting timeliens. Furthermore, I would engineer the prompt better so that the model has the ability to make logical conclusions with regards to soft skills. For instance, because the LLM is acting as a recruiter, it can make inferences about an applicants work ethic, leadership, or collaboration skills based on a strong professional job history. Alternatively, the model can ignore soft skills as a list of keywords entirely, making the application of the job fit solely technical, or hard skill focused.
